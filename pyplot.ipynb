1. Given the text below, create a bar plot (or a horizontal bar plot) that shows the frequency distribution of characters in the string. 
A dictionary has been created for you containing the characters and the frequencies of the characters.
Make sure your bar plot has the bars labelled correctly (with the character it represents the frequency of).


```python
# Import necessary libraries
import matplotlib.pyplot as plt
from collections import Counter

# Given text
TEXT = """At three o’clock precisely I was at Baker Street, but Holmes had not yet returned. 
          The landlady informed me that he had left the house shortly after eight o’clock in the morning. 
          I sat down beside the fire, however, with the intention of awaiting him, however long he might be. 
          I was already deeply interested in his inquiry, for, though it was surrounded by none of the grim and 
          strange features which were associated with the two crimes which I have already recorded, still, the nature 
          of the case and the exalted station of his client gave it a character of its own. Indeed, apart from the 
          nature of the investigation which my friend had on hand, there was something in his masterly grasp of a 
          situation, and his keen, incisive reasoning, which made it a pleasure to me to study his system of work, and 
          to follow the quick, subtle methods by which he disentangled the most inextricable mysteries. So accustomed 
          was I to his invariable success that the very possibility of his failing had ceased to enter into my head.""" 

# Convert all text to lowercase and count character frequencies
letter_freq = dict(Counter(TEXT.lower()))

# Create a bar plot for the character frequencies
plt.figure(figsize=(10, 6))
plt.bar(letter_freq.keys(), letter_freq.values(), color='skyblue')

# Add labels and title
plt.xlabel('Characters')
plt.ylabel('Frequency')
plt.title('Character Frequency Distribution')

# Rotate x-axis labels for readability
plt.xticks(rotation=90)

# Display the plot
plt.tight_layout()  # Adjust the layout to make it look nicer
plt.show()
```

2. Generate a sequence of at least 5 random numbers (integers are fine), and visualize the data using both `scatter` and `plot` on different axes.  You may use the index of the value as its x value in the plot.

```python
import matplotlib.pyplot as plt
import random

# Step 1: Generate a sequence of at least 5 random integers
random_numbers = [random.randint(1, 100) for _ in range(5)]

# Step 2: Define the x-values (index of the random numbers)
x_values = list(range(len(random_numbers)))

# Step 3: Create subplots for both scatter and line plot
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Scatter Plot on the first axis (ax1)
ax1.scatter(x_values, random_numbers, color='blue')
ax1.set_title('Scatter Plot')
ax1.set_xlabel('Index')
ax1.set_ylabel('Random Numbers')

# Line Plot on the second axis (ax2)
ax2.plot(x_values, random_numbers, color='green', marker='o')
ax2.set_title('Line Plot')
ax2.set_xlabel('Index')
ax2.set_ylabel('Random Numbers')

# Adjust the layout for better spacing between the plots
plt.tight_layout()

# Display the plots
plt.show()
```

3. Do the same question as above, but change the syle and color of both the scatter and the plot.

```python
import matplotlib.pyplot as plt
import random

# Step 1: Generate a sequence of at least 5 random integers
random_numbers = [random.randint(1, 100) for _ in range(5)]

# Step 2: Define the x-values (index of the random numbers)
x_values = list(range(len(random_numbers)))

# Step 3: Create subplots for both scatter and line plot with different styles and colors
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

# Scatter Plot with customized style and color
ax1.scatter(x_values, random_numbers, color='red', marker='D', s=100)  # Red color, diamond marker, size 100
ax1.set_title('Custom Styled Scatter Plot')
ax1.set_xlabel('Index')
ax1.set_ylabel('Random Numbers')

# Line Plot with customized style and color
ax2.plot(x_values, random_numbers, color='blue', linestyle='--', marker='o', markersize=8)  # Blue color, dashed line, circle markers
ax2.set_title('Custom Styled Line Plot')
ax2.set_xlabel('Index')
ax2.set_ylabel('Random Numbers')

# Adjust the layout for better spacing between the plots
plt.tight_layout()

# Display the plots
plt.show()
```

4. Given the data sets below, plot or scatter both on the same set of axes.  You are plotting the time it takes to execute a sorting algorithm on a list of a given size; your x axis should be the size, and the y axis is time.  Assume that `insertion_sort_times[i]` is how long it took to sort a list of length `sizes[i]`.  Add a legend to your plot showing which is insertion sort and which is merge sort.  Use the variable names to determine which data is which. 

```python
import matplotlib.pyplot as plt

# Data
sizes = [10, 20, 50, 100, 1000, 5000]
insertion_sort_times = [0.0010252999999999998, 0.0027026999999999954, 0.010147200000000002, 0.0381137, 3.6303399, 91.2180796]
merge_sort_times = [0.00161889999999687, 0.003635600000009731, 0.0090655000000055, 0.020108000000007564, 0.2687474000000094, 1.6147050999999948]

# Plotting both sorting times on the same axes
plt.plot(sizes, insertion_sort_times, label='Insertion Sort', color='red', marker='o', linestyle='-', markersize=6)
plt.plot(sizes, merge_sort_times, label='Merge Sort', color='blue', marker='x', linestyle='--', markersize=6)

# Adding titles and labels
plt.title('Sorting Algorithm Time Complexity')
plt.xlabel('List Size')
plt.ylabel('Execution Time (seconds)')

# Adding a legend to differentiate between the algorithms
plt.legend()

# Display the plot
plt.show()
```

### 5. Compare Multiple Data Sets Using Subplots

Given two data sets, create subplots to compare their distributions. One of the subplots should display a scatter plot, and the other should display a line plot of the data. Use random data sets to visualize and compare them.

```python
import numpy as np

# Generate random data for example
data1 = np.random.randint(1, 100, size=10)
data2 = np.random.randint(1, 100, size=10)

# Print the generated random data
print("Data1:", data1)
print("Data2:", data2)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data provided
data1 = np.array([58, 43, 3, 67, 21, 56, 2, 57, 35, 80])
data2 = np.array([92, 7, 44, 63, 39, 41, 16, 54, 85, 48])

# Create subplots with 1 row and 2 columns
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# Scatter plot on the first subplot (ax1)
ax1.scatter(range(len(data1)), data1, color='blue', label='Data1')
ax1.set_xlabel('Index')
ax1.set_ylabel('Value')
ax1.legend()

# Line plot on the second subplot (ax2)
ax2.plot(range(len(data2)), data2, color='red', label='Data2')
ax2.set_xlabel('Index')
ax2.set_ylabel('Value')
ax2.legend()

# Add a single title for the entire figure
fig.suptitle('Comparison of Data1 (Scatter Plot) and Data2 (Line Plot)', fontsize=16)

# Adjust layout to avoid overlapping content and show the plot
plt.tight_layout(rect=[0, 0, 1, 0.96])  # Adjust the title positioning

plt.show()
```
